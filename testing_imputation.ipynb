```python
cat_col = ['sex','embarked','class','who','adult_male','deck', 'embark_town', 'alive', 'alone']
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import KNNImputer
ode = OrdinalEncoder()
KNN_imp = KNNImputer()

def encodeCol(col):
    not_null = col.loc[col.notna()]
    temp_encoded = ode.fit_transform(not_null.values.reshape(-1,1))
    col.loc[col.notna()] = np.squeeze(temp_encoded)
    return col
```

```python
for columns in cat_col:
    df[columns] = encodeCol(df[columns])
```

```python
df
```

```python
pd.DataFrame(KNN_imp.fit_transform(df), columns=df.columns)
```
